# Task 11: Principal Component Analysis (PCA)

## Dataset: Wine Dataset

### Objective

- Load the Wine Dataset
- Explore the dataset
- Standardize the features
- Apply Principal Component Analysis (PCA)
- Reduce high-dimensional data into 2 principal components
- Analyze explained variance ratio
- Visualize transformed dataset in 2D

### Tools Used

- Python
- Pandas
- NumPy
- Scikit-Learn
- Matplotlib
- Google Colab

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.datasets import load_wine

from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA
```

# Load Wine Dataset

```python
wine = load_wine()

X = wine.data

y = wine.target
```

# Convert Dataset to DataFrame

```python
df = pd.DataFrame(
    X,
    columns=wine.feature_names
)

df["target"] = y

df.head()
```

# Dataset Exploration

```python
print("Dataset Shape:", df.shape)
```

```python
df.info()
```

```python
df.describe()
```

```python
print(wine.target_names)
```

# Feature Selection

```python
X = df.drop("target", axis=1)

y = df["target"]
```

```python
# Feature Scaling
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
# Apply Principal Component Analysis
```

```python
pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)
```

```python
# Transformed Dataset
```

```python
pca_df = pd.DataFrame(
    X_pca,
    columns=["Principal Component 1", "Principal Component 2"]
)

pca_df["target"] = y

pca_df.head()
```

# Explained Variance Ratio

```python
print(
    "Explained Variance Ratio:"
)

print(
    pca.explained_variance_ratio_
)
```

# Variance Plot

```python
plt.figure(figsize=(6,4))

plt.bar(
    ["PC1", "PC2"],
    pca.explained_variance_ratio_
)

plt.title("Explained Variance Ratio")

plt.ylabel("Variance")

plt.show()
```

# PCA Visualization

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X_pca[:,0],
    X_pca[:,1],
    c=y
)

plt.xlabel("Principal Component 1")

plt.ylabel("Principal Component 2")

plt.title("PCA - Wine Dataset")

plt.show()
```

# Observation

- PCA successfully reduced 13 numerical features into 2 principal components.
- The transformed data preserves most of the important information while reducing dimensionality.
- The visualization shows how different wine classes are distributed in a lower-dimensional space.

# Conclusion

Principal Component Analysis (PCA) was successfully applied to the Wine Dataset.

The dataset was standardized before dimensionality reduction to ensure equal contribution of all features.

PCA reduced the original feature space to two principal components while preserving a significant amount of variance.

The explained variance ratio and 2D visualization demonstrated the effectiveness of PCA in simplifying high-dimensional data for analysis and visualization.
